```python
import torch
from torch import nn, optim
import torch.nn.functional as F

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.18)
print("size of X_train = {} \nsize of X_test = {}".format(len(x_train),len(x_test)))

X_train = torch.from_numpy(x_train.to_numpy()).float()
y_train = torch.squeeze(torch.from_numpy(y_train.to_numpy()).float())
X_test = torch.from_numpy(x_test.to_numpy()).float()
y_test = torch.squeeze(torch.from_numpy(y_test.to_numpy()).float())

print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python

class Net(nn.Module):
    def __init__(self, n_features):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(n_features, 32)
        self.fc2 = nn.Linear(32, 8)
        self.dropout = nn.Dropout(0.5,inplace = True)
        self.fc3 = nn.Linear(8, 1)
        
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        output = torch.sigmoid(self.fc3(x))
        return output
    
net = Net(X_train.shape[1]) # guess what I mean by no_features, what do you think the input nodes are
```

```python
net
```

```python
criterion = nn.BCELoss()
optimizer = optim.Adam(net.parameters(), lr=0.001)
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print("using ",device)
net = net.to(device)
criterion = criterion.to(device)

# since we are not using batches, we will put data in device directly outsie train loop
X_train = X_train.to(device)
y_train = y_train.to(device)
X_test = X_test.to(device)
y_test = y_test.to(device)
```

```python
def calculate_accuracy(y_true, y_pred):
    predicted = y_pred.ge(.5).view(-1)
    return (y_true == predicted).sum().float() / len(y_true)

for epoch in range(1000):
    
    y_pred = net(X_train)
    y_pred = torch.squeeze(y_pred)
    train_loss = criterion(y_pred, y_train)

    if epoch %10 == 0:
        train_acc = calculate_accuracy(y_train, y_pred)
        y_test_pred = net(X_test)
        y_test_pred = torch.squeeze(y_test_pred)
        test_loss = criterion(y_test_pred, y_test)
        test_acc = calculate_accuracy(y_test, y_test_pred)
        print('epoch {} Train set - loss: {}, accuracy: {} Test  set - loss: {}, accuracy: {}'.format(epoch,train_loss,train_acc,test_loss,test_acc))

    optimizer.zero_grad()
    train_loss.backward()
    optimizer.step()
    
print('Final epoch {} Train set - loss: {}, accuracy: {} Test  set - loss: {}, accuracy: {}'.format(epoch,train_loss,train_acc,test_loss,test_acc))
```
